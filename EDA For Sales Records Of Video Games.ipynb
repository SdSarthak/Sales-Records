```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('vgsales.csv')
```

```python
df=df.dropna()
```

```python
df['Year'].hist()
plt.title('Distribution of Release Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['NA_Sales'])
plt.title('Box Plot of NA Sales')
sns.boxplot(x=df['EU_Sales'])
plt.title('Box Plot of EU Sales')
sns.boxplot(x=df['JP_Sales'])
plt.title('Box Plot of JP Sales')
sns.boxplot(x=df['Other_Sales'])
plt.title('Box Plot of Other Sales')
sns.boxplot(x=df['Blobal_Sales'])
plt.title('Box Plot of Global Sales')
plt.show()
```

```python
df['Genre'].value_counts().plot(kind='bar')
plt.title('Number of Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
df['Platform'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Market Share by Platform')
plt.show()
```

```python
plt.scatter(df['NA_Sales'], df['Global_Sales'])
plt.title('Global Sales vs NA Sales')
plt.xlabel('NA Sales')
plt.ylabel('Global Sales')
plt.show()
```

```python
sns.pairplot(df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']])
plt.show()
```

```python
correlation_matrix = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.groupby('Year')['Global_Sales'].sum().plot(kind='line')
plt.title('Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Total Global Sales')
plt.show()
```

```python
sns.countplot(x='Genre', data=df)
plt.title('Count of Games by Genre')
plt.show()
```
